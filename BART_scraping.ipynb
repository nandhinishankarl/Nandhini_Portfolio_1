```python
# scraping information on all the BART routes using the publicly available API
requestString = 'https://api.bart.gov/api/route.aspx?cmd=routeinfo&route=all&date=now&key=MW9S-E7SL-26DU-VV8V&json=y'
r_route = requests.get(requestString)
```

```python
import pandas as pd
```

```python
# converting the dictionary to a data frame
# route information along with their names, ids and stations can be found in the table
df_routes = pd.DataFrame(d_route['root']['routes']['route'])
df_routes
```

```python
# Extracting the station names from the config column
df_routes["stations"] = df_routes["config"].apply(lambda x: x["station"])

# Flattening station names into individual rows
stations_expanded = df_routes.explode("stations")

print(stations_expanded)
```

```python
stations_expanded.head()
```

```python
# creating a column called order of points to get the order of stations in each route
stations_expanded['order_of_points'] = stations_expanded.groupby("name").cumcount() + 1
```

```python
file_path = '/content/BART_stations.xlsx'
df_stations = pd.read_excel(file_path)
print(df_stations.head())
```

```python
# Step 1: Explode the `stations` column in stations_expanded to create one row per station abbreviation
stations_expanded_exploded = stations_expanded.explode('stations')

# Step 2: Merge the exploded dataframe with df_stations using the station abbreviations
# The goal is to join both tables so location based information like latitude and longitude coordinates can be used
# for the station values on the routes
merged_df = stations_expanded_exploded.merge(
    df_stations,
    left_on='stations',
    right_on='abbr',
    how='left'
)

# Step 4: Check the resulting dataframe
print(merged_df.head())
```
